```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the weather data from a CSV file into a Pandas DataFrame
weather_data = pd.read_csv('/content/Weather Data.csv')
```

```python
# Display the first few rows of the DataFrame to understand the structure of the data
print(weather_data.head())
```

```python
# Describe the data to get statistical information about numerical columns
print(weather_data.describe())
```

```python
# Assuming your DataFrame has a DateTime index, and 'Temp_C' represents temperature
plt.figure(figsize=(10, 6))
plt.plot(weather_data.index, weather_data['Temp_C'])
plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print(weather_data.columns)
```

```python
# Assuming 'weather_data' is your DataFrame
plt.figure(figsize=(12, 8))

# Loop through each column (excluding non-numeric columns like 'Date/Time' and 'Weather')
for column in weather_data.columns:
    if weather_data[column].dtype in ['int64', 'float64']:  # Check if the column is numeric
        sns.histplot(weather_data[column], bins=20, kde=True)
        plt.title(f'{column} Distribution')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```

```python
# Visualizing wind speed distribution using a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=weather_data['Wind Speed_km/h'])
plt.title('Wind Speed Distribution')
plt.ylabel('Wind Speed')
plt.show()
```
